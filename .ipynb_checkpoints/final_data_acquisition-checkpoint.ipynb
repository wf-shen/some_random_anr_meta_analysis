```python
import requests
import json
import time
import pandas as pd # å¯¼å…¥pandasåº“ï¼Œé€šå¸¸ç®€å†™ä¸ºpd
import os
from datetime import date, timedelta # å¯¼å…¥æ—¥æœŸæ—¶é—´å·¥å…·ï¼Œæ–¹ä¾¿å¤„ç†æ—¥æœŸ
```

```python
import requests
from datetime import date, timedelta

def fetch_tournaments(start_date, end_date, game_format='standard'):
    """
    ä»AlwaysBeRunning APIè·å–æŒ‡å®šæ—¶é—´èŒƒå›´å’Œèµ›åˆ¶ä¸‹çš„é”¦æ ‡èµ›æ•°æ®ã€‚
    ï¼ˆå·²ä¿®æ­£å‚æ•°é”®åï¼‰
    """
    print(f"\n--- æ­£åœ¨è·å–æ•°æ® ---")
    print(f"æ—¶é—´èŒƒå›´: {start_date} åˆ° {end_date}, èµ›åˆ¶: {game_format}")
    
    # ä¼˜åŒ–ç‚¹ï¼šbase_urlæœ«å°¾ä¸éœ€è¦ '?'
    base_url = "https://alwaysberunning.net/api/tournaments" 
    
    # å…³é”®ä¿®æ­£ï¼šå°†é”®åä¸¥æ ¼ä¿®æ”¹ä¸ºAPIæ–‡æ¡£è¦æ±‚çš„ 'start' å’Œ 'end'
    params = {
        'format': game_format,
        'concluded': 1,
        'start': start_date,  # <-- ä¿®æ­£è¿™é‡Œ
        'end': end_date      # <-- ä¿®æ­£è¿™é‡Œ
    }
    
    try:
        response = requests.get(base_url, params=params)
        
        # æˆ‘ä»¬å¯ä»¥æ‰“å°å‡ºæœ€ç»ˆURLæ¥éªŒè¯ä¸€ä¸‹
        print(f"å®é™…è¯·æ±‚çš„URL: {response.url}")
        
        response.raise_for_status()
        tournaments = response.json()
        print(f"âœ… æˆåŠŸè·å–äº† {len(tournaments)} åœºæ¯”èµ›çš„æ•°æ®ã€‚")
        return tournaments
    except requests.exceptions.RequestException as e:
        print(f"âŒ è¯·æ±‚å¤±è´¥: {e}")
        return []

# --- ç°åœ¨å†æ¬¡è°ƒç”¨å‡½æ•° ---

# è·å–2025å¹´8æœˆ1æ—¥åˆ°9æœˆ17æ—¥çš„æ•°æ®
standard_tournaments_2508 = fetch_tournaments(
    start_date='2025.07.18',
    end_date='2025.09.17',
    game_format='standard'
)

# ç°åœ¨è¿™é‡Œçš„æ•°é‡åº”è¯¥ä¼šå˜å›ä½ é¢„æœŸçš„å‡ ç™¾åœºäº†
print(f"\næœ€ç»ˆæ‹¿åˆ° {len(standard_tournaments_2508)} æ¡æ¯”èµ›è®°å½•ã€‚")

# æŠŠæ¯”èµ›æ•°æ®èµ‹å€¼ç»™tournaments_list
tournaments_list = standard_tournaments_2508

if tournaments_list:
    print("\næ•°æ®è·å–å’Œèµ‹å€¼æˆåŠŸï¼")
else:
    print("\næ²¡æœ‰æ‰¾åˆ°ç¬¦åˆæ¡ä»¶çš„æ¯”èµ›è®°å½•ã€‚")
```

```python
# --- æ•°æ®ç­›é€‰æ­¥éª¤ ---

# 1. åˆ›å»ºä¸€ä¸ªç©ºåˆ—è¡¨ï¼Œç”¨æ¥å­˜æ”¾ç¬¦åˆæˆ‘ä»¬æ¡ä»¶çš„é”¦æ ‡èµ›ID
filtered_ids = []

print(f"å¼€å§‹ä» {len(tournaments_list)} åœºé”¦æ ‡èµ›ä¸­è¿›è¡Œç­›é€‰...")
print("ç­›é€‰æ¡ä»¶: claim_count > 3 AND claim_conflict is False\n")

# 2. éå†ä½ è·å–åˆ°çš„æ¯ä¸€åœºé”¦æ ‡èµ›æ•°æ®
for tournament in tournaments_list:
    
    # 3. æ£€æŸ¥æ¯ä¸€ä¸ªé”¦æ ‡èµ›å­—å…¸æ˜¯å¦åŒæ—¶æ»¡è¶³ä¸¤ä¸ªæ¡ä»¶
    # ä½¿ç”¨ .get() æ–¹æ³•æ¥å®‰å…¨åœ°è·å–å€¼ï¼Œå¦‚æœé”®ä¸å­˜åœ¨ï¼Œåˆ™è¿”å›ä¸€ä¸ªä¸ä¼šå¯¼è‡´é”™è¯¯çš„å€¼
    # ä¾‹å¦‚ï¼Œå¦‚æœ'claim_count'ä¸å­˜åœ¨ï¼Œæˆ‘ä»¬ç»™å®ƒä¸€ä¸ªé»˜è®¤å€¼0
    claim_count = tournament.get('claim_count', 0)
    claim_conflict = tournament.get('claim_conflict', True) # å¦‚æœ'claim_conflict'ä¸å­˜åœ¨ï¼Œé»˜è®¤å®ƒæ˜¯æœ‰é—®é¢˜çš„

    # è¿™å°±æ˜¯ç­›é€‰çš„æ ¸å¿ƒé€»è¾‘ï¼
    if claim_count > 3 and claim_conflict == False:
        # 4. å¦‚æœæ¡ä»¶æ»¡è¶³ï¼Œå°±æŠŠè¿™åœºé”¦æ ‡èµ›çš„'id'æ·»åŠ åˆ°æˆ‘ä»¬çš„åˆ—è¡¨ä¸­
        filtered_ids.append(tournament['id'])
        # æˆ‘ä»¬å¯ä»¥é¡ºä¾¿æ‰“å°å‡ºç¬¦åˆæ¡ä»¶çš„æ¯”èµ›ä¿¡æ¯ï¼Œæ–¹ä¾¿æ£€æŸ¥
        print(f"âœ… ç¬¦åˆæ¡ä»¶: ID={tournament['id']}, Title='{tournament['title']}', claim_count={claim_count}")

# 5. å¾ªç¯ç»“æŸåï¼Œæ‰“å°å‡ºæ‰€æœ‰ç¬¦åˆæ¡ä»¶çš„ID
print("\n" + "="*50)
print("ğŸ‰ ç­›é€‰å®Œæˆï¼")

if filtered_ids:
    print(f"æ€»å…±æœ‰ {len(filtered_ids)} åœºé”¦æ ‡èµ›ç¬¦åˆæ¡ä»¶ã€‚")
    print("å®ƒä»¬çš„IDåˆ†åˆ«æ˜¯:")
    print(filtered_ids)
else:
    print("æ²¡æœ‰æ‰¾åˆ°ç¬¦åˆæ¡ä»¶çš„é”¦æ ‡èµ›ã€‚")
```

```python
# tournaments_list = response.json() 

# --- æ•°æ®ç­›é€‰æ­¥éª¤ ---

# 1. åˆ›å»ºä¸€ä¸ªç©ºåˆ—è¡¨ï¼Œç”¨æ¥å­˜æ”¾ç¬¦åˆæˆ‘ä»¬æ¡ä»¶çš„é”¦æ ‡èµ›ID
filtered_ids = []

print(f"å¼€å§‹ä» {len(tournaments_list)} åœºé”¦æ ‡èµ›ä¸­è¿›è¡Œç­›é€‰...")
print("ç­›é€‰æ¡ä»¶: claim_count > 3 AND claim_conflict is False\n")

# 2. éå†ä½ è·å–åˆ°çš„æ¯ä¸€åœºé”¦æ ‡èµ›æ•°æ®
for tournament in tournaments_list:
    
    # 3. æ£€æŸ¥æ¯ä¸€ä¸ªé”¦æ ‡èµ›å­—å…¸æ˜¯å¦åŒæ—¶æ»¡è¶³ä¸¤ä¸ªæ¡ä»¶
    # ä½¿ç”¨ .get() æ–¹æ³•æ¥å®‰å…¨åœ°è·å–å€¼ï¼Œå¦‚æœé”®ä¸å­˜åœ¨ï¼Œåˆ™è¿”å›ä¸€ä¸ªä¸ä¼šå¯¼è‡´é”™è¯¯çš„å€¼
    # ä¾‹å¦‚ï¼Œå¦‚æœ'claim_count'ä¸å­˜åœ¨ï¼Œæˆ‘ä»¬ç»™å®ƒä¸€ä¸ªé»˜è®¤å€¼0
    claim_count = tournament.get('claim_count', 0)
    claim_conflict = tournament.get('claim_conflict', True) # å¦‚æœ'claim_conflict'ä¸å­˜åœ¨ï¼Œé»˜è®¤å®ƒæ˜¯æœ‰é—®é¢˜çš„

    # è¿™å°±æ˜¯ç­›é€‰çš„æ ¸å¿ƒé€»è¾‘ï¼
    if claim_count > 3 and claim_conflict == False:
        # 4. å¦‚æœæ¡ä»¶æ»¡è¶³ï¼Œå°±æŠŠè¿™åœºé”¦æ ‡èµ›çš„'id'æ·»åŠ åˆ°æˆ‘ä»¬çš„åˆ—è¡¨ä¸­
        filtered_ids.append(tournament['id'])
        # æˆ‘ä»¬å¯ä»¥é¡ºä¾¿æ‰“å°å‡ºç¬¦åˆæ¡ä»¶çš„æ¯”èµ›ä¿¡æ¯ï¼Œæ–¹ä¾¿æ£€æŸ¥
        print(f"âœ… ç¬¦åˆæ¡ä»¶: ID={tournament['id']}, Title='{tournament['title']}', claim_count={claim_count}")

# 5. å¾ªç¯ç»“æŸåï¼Œæ‰“å°å‡ºæ‰€æœ‰ç¬¦åˆæ¡ä»¶çš„ID
print("\n" + "="*50)
print("ğŸ‰ ç­›é€‰å®Œæˆï¼")

if filtered_ids:
    print(f"æ€»å…±æœ‰ {len(filtered_ids)} åœºé”¦æ ‡èµ›ç¬¦åˆæ¡ä»¶ã€‚")
    print("å®ƒä»¬çš„IDåˆ†åˆ«æ˜¯:")
    print(filtered_ids)
else:
    print("æ²¡æœ‰æ‰¾åˆ°ç¬¦åˆæ¡ä»¶çš„é”¦æ ‡èµ›ã€‚")
```

```python
# 1. åˆ›å»ºä¸¤ä¸ªç©ºåˆ—è¡¨ï¼Œåˆ†åˆ«å­˜æ”¾å…¬å¸å’Œæ½œè¢­è€…çš„URLä¿¡æ¯
all_corp_deck_urls = []
all_runner_deck_urls = []

print(f"å‡†å¤‡å¼€å§‹ä» {len(filtered_ids)} åœºç­›é€‰åçš„é”¦æ ‡èµ›ä¸­è·å–ç‰Œè¡¨é“¾æ¥...\n")

# 2. éå†æ¯ä¸€ä¸ªç­›é€‰å‡ºæ¥çš„é”¦æ ‡èµ›ID
for tournament_id in filtered_ids:
    
    # æ„å»ºé’ˆå¯¹å•ä¸ªé”¦æ ‡èµ›çš„API URLå’Œå‚æ•°
    params = {'id': tournament_id}
    url = "https://alwaysberunning.net/api/entries"
    
    print(f"æ­£åœ¨è¯·æ±‚é”¦æ ‡èµ› ID: {tournament_id} çš„å‚èµ›åˆ—è¡¨...")
    
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        
        # 3. response.json() è¿”å›çš„æ˜¯ä¸€ä¸ªå‚èµ›è€…åˆ—è¡¨ (list of entries)
        # æˆ‘ä»¬ç›´æ¥æŠŠå®ƒå‘½åä¸º entries_list
        entries_list = response.json()
        
        # 4. æ£€æŸ¥è¿™ä¸ªå‚èµ›è€…åˆ—è¡¨æ˜¯å¦ä¸ºç©º
        if entries_list:
            print(f"  -> æˆåŠŸè·å–äº† {len(entries_list)} æ¡å‚èµ›è®°å½•ã€‚")
            
            # 5. éå†è¿™åœºæ¯”èµ›ä¸­çš„ã€æ¯ä¸€æ¡å‚èµ›è®°å½•ã€‘(æ¯ä¸€ä¸ªå‚èµ›è€…å­—å…¸)
            for entry in entries_list:
                
                # 6. ä»æ¯ä¸ªå‚èµ›è€…(entry)å­—å…¸ä¸­æå–URL
                corp_url = entry.get('corp_deck_url')
                runner_url = entry.get('runner_deck_url')
                player_name = entry.get('user_name', 'N/A')
                
                # 7. å¦‚æœå…¬å¸URLå­˜åœ¨ï¼Œå°±æŠŠå®ƒå­˜åˆ°å…¬å¸åˆ—è¡¨ä¸­
                if corp_url:
                    all_corp_deck_urls.append({
                        'tournament_id': tournament_id,
                        'player': player_name,
                        'corp_deck_url': corp_url
                    })
                
                # 8. å¦‚æœæ½œè¢­è€…URLå­˜åœ¨ï¼Œå°±æŠŠå®ƒå­˜åˆ°æ½œè¢­è€…åˆ—è¡¨ä¸­
                if runner_url:
                    all_runner_deck_urls.append({
                        'tournament_id': tournament_id,
                        'player': player_name,
                        'runner_deck_url': runner_url
                    })
        else:
            print(f"  -> é”¦æ ‡èµ› {tournament_id} æ²¡æœ‰è¿”å›å‚èµ›è®°å½•ã€‚")

    except requests.exceptions.RequestException as e:
        print(f"  -> é”™è¯¯: è¯·æ±‚é”¦æ ‡èµ› {tournament_id} å¤±è´¥: {e}")

    # ç¤¼è²Œåœ°æš‚åœä¸€ç§’
    time.sleep(1)

# --- å¾ªç¯ç»“æŸåï¼ŒæŸ¥çœ‹æˆ‘ä»¬çš„æˆæœ ---
print("\n" + "="*50)
print("ğŸ‰ å…¨éƒ¨å¤„ç†å®Œæˆï¼")

# å…¬å¸URLè¾“å‡º
if all_corp_deck_urls:
    print(f"\næ€»å…±æ”¶é›†åˆ°äº† {len(all_corp_deck_urls)} ä»½å…¬å¸ç‰Œè¡¨é“¾æ¥ã€‚")
    print("å‰5æ¡è®°å½•ç¤ºä¾‹ï¼š")
    print(json.dumps(all_corp_deck_urls[:5], indent=2))
else:
    print("\næœªèƒ½æ”¶é›†åˆ°ä»»ä½•å…¬å¸ç‰Œè¡¨é“¾æ¥ã€‚")

# æ½œè¢­è€…URLè¾“å‡º
if all_runner_deck_urls:
    print(f"\næ€»å…±æ”¶é›†åˆ°äº† {len(all_runner_deck_urls)} ä»½æ½œè¢­è€…ç‰Œè¡¨é“¾æ¥ã€‚")
    print("å‰5æ¡è®°å½•ç¤ºä¾‹ï¼š")
    print(json.dumps(all_runner_deck_urls[:5], indent=2))
else:
    print("\næœªèƒ½æ”¶é›†åˆ°ä»»ä½•æ½œè¢­è€…ç‰Œè¡¨é“¾æ¥ã€‚")
```

```python

# --- ç¬¬ä¸€éƒ¨åˆ†ï¼šå¤„ç†å…¬å¸æ–¹å¡ç»„ ---

# 1. åˆ›å»ºä¸€ä¸ªæœ€ç»ˆçš„ã€å¤§çš„åˆ—è¡¨ï¼Œç”¨æ¥å­˜æ”¾æ¯ä¸€æ¡å¡ç‰Œè®°å½•
corp_card_data = []

print(f"å‡†å¤‡å¤„ç† {len(all_corp_deck_urls)} ä¸ªå…¬å¸æ–¹å¡ç»„é“¾æ¥...\n")

# 2. éå†æˆ‘ä»¬ä¹‹å‰æ”¶é›†åˆ°çš„æ¯ä¸€ä¸ªåŒ…å«URLçš„å­—å…¸
for deck_info in all_corp_deck_urls:
    
    url = deck_info['corp_deck_url']
    
    # å¦‚æœURLä¸ºç©ºï¼Œåˆ™è·³è¿‡æ­¤æ¬¡å¾ªç¯
    if not url:
        continue

    try:
        # 3. ä»URLä¸­è§£æå‡º decklist_id (è¿™æ˜¯å…³é”®ä¸€æ­¥ï¼)
        # ä¾‹å¦‚: 'https://netrunnerdb.com/en/decklist/91006/...' -> '91006'
        decklist_id = url.split('/')[5]
        
        # 4. ä½¿ç”¨IDè°ƒç”¨NetrunnerDB API
        api_url = f"https://netrunnerdb.com/api/2.0/public/decklist/{decklist_id}"
        print(f"æ­£åœ¨è¯·æ±‚ Decklist ID: {decklist_id} ...")
        
        response = requests.get(api_url)
        response.raise_for_status()
        decklist_data = response.json()
        
        # 5. æå– 'cards' å­—å…¸
        if decklist_data.get('success') and decklist_data.get('data'):
            cards_dict = decklist_data['data'][0].get('cards')
            
            if cards_dict:
                # 6. ã€æ ¸å¿ƒã€‘éå† 'cards' å­—å…¸ï¼Œå°†æ¯å¼ å¡ä½œä¸ºä¸€æ¡æ–°è®°å½•æ·»åŠ 
                for card_id, quantity in cards_dict.items():
                    corp_card_data.append({
                        'tournament_id': deck_info['tournament_id'],
                        'player': deck_info['player'],
                        'decklist_id': decklist_id,
                        'card_id': card_id,
                        'quantity': quantity
                    })
        
        # ç¤¼è²Œåœ°æš‚åœ
        time.sleep(0.5)

    except Exception as e:
        print(f"  -> å¤„ç†URL {url} æ—¶å‘ç”Ÿé”™è¯¯: {e}")
        continue # å³ä½¿å•ä¸ªURLå‡ºé”™ï¼Œä¹Ÿç»§ç»­å¤„ç†ä¸‹ä¸€ä¸ª

```

```python
# --- ç¬¬äºŒéƒ¨åˆ†ï¼šå°†æ”¶é›†åˆ°çš„æ•°æ®è½¬æ¢æˆè¡¨æ ¼ (Pandas DataFrame) ---
print("\n" + "="*50)
print("ğŸ‰ å…¬å¸æ–¹æ•°æ®å¤„ç†å®Œæˆï¼æ­£åœ¨åˆ›å»ºè¡¨æ ¼...")

if corp_card_data:
    # è¿™å°±æ˜¯â€œæ‹¼æˆè¡¨æ ¼â€çš„é­”æ³•ï¼
    corp_df = pd.DataFrame(corp_card_data)

    print("\nå…¬å¸æ–¹å¡ç‰Œæ•°æ®è¡¨æ ¼åˆ›å»ºæˆåŠŸï¼")
    print("è¡¨æ ¼ä¿¡æ¯æ¦‚è§ˆï¼š")
    corp_df.info() # æ˜¾ç¤ºè¡¨æ ¼çš„ç»“æ„ä¿¡æ¯

    print("\nè¡¨æ ¼å‰5è¡Œé¢„è§ˆï¼š")
    print(corp_df.head()) # æ˜¾ç¤ºè¡¨æ ¼çš„å‰5è¡Œ
else:
    print("æœªèƒ½æ”¶é›†åˆ°ä»»ä½•å…¬å¸æ–¹å¡ç‰Œæ•°æ®ã€‚")
```

```python

# --- ç¬¬äºŒéƒ¨åˆ†ï¼šå¤„ç†æ½œè¢­è€…æ–¹å¡ç»„ ---

# 1. åˆ›å»ºä¸€ä¸ªæœ€ç»ˆçš„ã€å¤§çš„åˆ—è¡¨ï¼Œç”¨æ¥å­˜æ”¾æ¯ä¸€æ¡å¡ç‰Œè®°å½•
runner_card_data = []

print(f"å‡†å¤‡å¤„ç† {len(all_runner_deck_urls)} ä¸ªæ½œè¢­è€…æ–¹å¡ç»„é“¾æ¥...\n")

# 2. éå†æˆ‘ä»¬ä¹‹å‰æ”¶é›†åˆ°çš„æ¯ä¸€ä¸ªåŒ…å«URLçš„å­—å…¸
for deck_info in all_runner_deck_urls:
    
    url = deck_info['runner_deck_url']
    
    # å¦‚æœURLä¸ºç©ºï¼Œåˆ™è·³è¿‡æ­¤æ¬¡å¾ªç¯
    if not url:
        continue

    try:
        # 3. ä»URLä¸­è§£æå‡º decklist_id (è¿™æ˜¯å…³é”®ä¸€æ­¥ï¼)
        # ä¾‹å¦‚: 'https://netrunnerdb.com/en/decklist/91006/...' -> '91006'
        decklist_id = url.split('/')[5]
        
        # 4. ä½¿ç”¨IDè°ƒç”¨NetrunnerDB API
        api_url = f"https://netrunnerdb.com/api/2.0/public/decklist/{decklist_id}"
        print(f"æ­£åœ¨è¯·æ±‚ Decklist ID: {decklist_id} ...")
        
        response = requests.get(api_url)
        response.raise_for_status()
        decklist_data = response.json()
        
        # 5. æå– 'cards' å­—å…¸
        if decklist_data.get('success') and decklist_data.get('data'):
            cards_dict = decklist_data['data'][0].get('cards')
            
            if cards_dict:
                # 6. ã€æ ¸å¿ƒã€‘éå† 'cards' å­—å…¸ï¼Œå°†æ¯å¼ å¡ä½œä¸ºä¸€æ¡æ–°è®°å½•æ·»åŠ 
                for card_id, quantity in cards_dict.items():
                    runner_card_data.append({
                        'tournament_id': deck_info['tournament_id'],
                        'player': deck_info['player'],
                        'decklist_id': decklist_id,
                        'card_id': card_id,
                        'quantity': quantity
                    })
        
        # ç¤¼è²Œåœ°æš‚åœ
        time.sleep(0.5)

    except Exception as e:
        print(f"  -> å¤„ç†URL {url} æ—¶å‘ç”Ÿé”™è¯¯: {e}")
        continue # å³ä½¿å•ä¸ªURLå‡ºé”™ï¼Œä¹Ÿç»§ç»­å¤„ç†ä¸‹ä¸€ä¸ª

```

```python
# --- ç¬¬äºŒéƒ¨åˆ†ï¼šå°†æ”¶é›†åˆ°çš„æ•°æ®è½¬æ¢æˆè¡¨æ ¼ (Pandas DataFrame) ---
print("\n" + "="*50)
print("ğŸ‰ æ½œè¢­è€…æ–¹æ•°æ®å¤„ç†å®Œæˆï¼æ­£åœ¨åˆ›å»ºè¡¨æ ¼...")

if runner_card_data:
    runner_df = pd.DataFrame(runner_card_data)

    print("\æ½œè¢­è€…æ–¹å¡ç‰Œæ•°æ®è¡¨æ ¼åˆ›å»ºæˆåŠŸï¼")
    print("è¡¨æ ¼ä¿¡æ¯æ¦‚è§ˆï¼š")
    runner_df.info() # æ˜¾ç¤ºè¡¨æ ¼çš„ç»“æ„ä¿¡æ¯

    print("\nè¡¨æ ¼å‰5è¡Œé¢„è§ˆï¼š")
    print(runner_df.head()) # æ˜¾ç¤ºè¡¨æ ¼çš„å‰5è¡Œ
else:
    print("æœªèƒ½æ”¶é›†åˆ°ä»»ä½•æ½œè¢­è€…æ–¹å¡ç‰Œæ•°æ®ã€‚")
```

```python
# --- å‡†å¤‡å·¥ä½œï¼šè·å–å¹¶å‡†å¤‡å¡ç‰Œä¿¡æ¯æ•°æ® ---

# 1. è·å–æ‰€æœ‰å¡ç‰Œçš„è¯¦ç»†ä¿¡æ¯
all_cards_url = "https://netrunnerdb.com/api/2.0/public/cards"
response = requests.get(all_cards_url)
all_cards_data = response.json()['data']

# 2. å°†å¡ç‰Œä¿¡æ¯åˆ—è¡¨è½¬æ¢æˆä¸€ä¸ªDataFrame
cards_info_df = pd.DataFrame(all_cards_data)

# 3. ä¸ºäº†è®©è¡¨æ ¼æ›´ç®€æ´ï¼Œæˆ‘ä»¬åªä¿ç•™éœ€è¦çš„å‡ åˆ—
# 'code'æ˜¯å¡ç‰ŒID, 'title'æ˜¯æ ‡é¢˜, 'faction_code'æ˜¯é˜µè¥, 'side_code'æ˜¯å…¬å¸/æ½œè¢­è€…
cards_info_df = cards_info_df[['code', 'title', 'faction_code', 'side_code']]

print("å¡ç‰Œä¿¡æ¯è¡¨æ ¼ï¼ˆcards_info_dfï¼‰å·²å‡†å¤‡å°±ç»ªï¼š")
print(cards_info_df.head())


# --- æ ¸å¿ƒæ­¥éª¤ï¼šåˆå¹¶è¡¨æ ¼ ---
print("\n" + "="*50)
print("æ­£åœ¨å°†å¡ç»„æ•°æ® (corp_df) ä¸å¡ç‰Œä¿¡æ¯ (cards_info_df) è¿›è¡Œåˆå¹¶...")

# ä½¿ç”¨ pd.merge() å‡½æ•°
# left_on='card_id' æŒ‡çš„æ˜¯ corp_df ä¸­ç”¨æ¥åŒ¹é…çš„åˆ—
# right_on='code' æŒ‡çš„æ˜¯ cards_info_df ä¸­ç”¨æ¥åŒ¹é…çš„åˆ—
corp_df_merged = pd.merge(corp_df, cards_info_df, left_on='card_id', right_on='code')

print("\nåˆå¹¶æˆåŠŸï¼æ–°çš„è¡¨æ ¼ corp_df_merged åŒ…å«äº†å¡ç‰Œçš„è¯¦ç»†ä¿¡æ¯ã€‚")
print("æ–°è¡¨æ ¼çš„å‰5è¡Œé¢„è§ˆï¼š")
print(corp_df_merged.head())
```

```python

# å‡è®¾ä½ å·²ç»æœ‰äº† runner_df å’Œ cards_info_df è¿™ä¸¤ä¸ªè¡¨æ ¼

print("="*50)
print("æ­£åœ¨å°†æ½œè¢­è€…æ–¹å¡ç»„æ•°æ® (runner_df) ä¸å¡ç‰Œä¿¡æ¯ (cards_info_df) è¿›è¡Œåˆå¹¶...")

# ä½¿ç”¨å®Œå…¨ç›¸åŒçš„ pd.merge() å‡½æ•°
# å·¦è¡¨æ˜¯ runner_dfï¼Œå³è¡¨æ˜¯ cards_info_df
# è¿æ¥é”®ä¾ç„¶æ˜¯ 'card_id' å’Œ 'code'
runner_df_merged = pd.merge(runner_df, cards_info_df, left_on='card_id', right_on='code')

print("\nåˆå¹¶æˆåŠŸï¼æ–°çš„è¡¨æ ¼ runner_df_merged å·²åˆ›å»ºã€‚")
print("æ–°è¡¨æ ¼çš„å‰5è¡Œé¢„è§ˆï¼š")
print(runner_df_merged.head())
```

```python
# --- æ•°æ®æŒä¹…åŒ–ï¼šä¿å­˜åˆ°CSVæ–‡ä»¶ ---

print("æ­£åœ¨å°†å¤„ç†å¥½çš„æ•°æ®ä¿å­˜åˆ°æœ¬åœ°æ–‡ä»¶...")

# ç¡®ä¿ä½ æœ‰ä¸€ä¸ªåä¸º'data'çš„å­æ–‡ä»¶å¤¹æ¥å­˜æ”¾æ•°æ®
import os
if not os.path.exists('data'):
    os.makedirs('data')

# to_csv() å°±æ˜¯ä¿å­˜å‘½ä»¤
# 'data/final_corp_data.csv' æ˜¯æ–‡ä»¶è·¯å¾„å’Œåç§°
# index=False æ˜¯ä¸€ä¸ªéå¸¸é‡è¦çš„å‚æ•°ï¼Œå®ƒèƒ½é˜²æ­¢PandasæŠŠDataFrameçš„è¡Œç´¢å¼•ä¹Ÿå­˜è¿›æ–‡ä»¶é‡Œ
corp_df_merged.to_csv('data/final_corp_data.csv', index=False,encoding='utf-8-sig')
runner_df_merged.to_csv('data/final_runner_data.csv', index=False,encoding='utf-8-sig')

print("âœ… æˆåŠŸï¼æ•°æ®å·²ä¿å­˜ä¸º data/final_corp_data.csv å’Œ data/final_runner_data.csv")
```
