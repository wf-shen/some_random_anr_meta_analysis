```python
import requests
import json

# --- ç¬¬ä¸€æ­¥ï¼šè·å–æ•°æ®å¹¶å­˜å…¥ tournaments_list ---

api_url = "https://alwaysberunning.net/api/tournaments?format=standard&concluded=1&start=2025.05.01&end=2025.08.30"

print(f"æ­£åœ¨å°è¯•ä»ä»¥ä¸‹åœ°å€è·å–æ•°æ®: {api_url}")

# ä½¿ç”¨requestsåº“çš„getæ–¹æ³•æ¥å‘é€ä¸€ä¸ªâ€œè¯·ç»™æˆ‘æ•°æ®â€çš„è¯·æ±‚
response = requests.get(api_url)

# æ£€æŸ¥è¯·æ±‚æ˜¯å¦æˆåŠŸã€‚çŠ¶æ€ç 200ä»£è¡¨â€œæˆåŠŸâ€
if response.status_code == 200:
    print("æˆåŠŸï¼æœåŠ¡å™¨å·²å“åº”ã€‚")
    
    # é­”æ³•å°±åœ¨è¿™ä¸€è¡Œï¼
    # response.json() è¿”å›äº†ä¸€ä¸ªåŒ…å«æ‰€æœ‰é”¦æ ‡èµ›ä¿¡æ¯çš„åˆ—è¡¨
    # æˆ‘ä»¬ç”¨'='æŠŠå®ƒèµ‹å€¼ç»™ä¸€ä¸ªæˆ‘ä»¬è‡ªå·±å‘½åçš„å˜é‡ï¼Œè¿™é‡Œå°±å« tournaments_list
    tournaments_list = response.json()
    
    print(f"æˆ‘ä»¬æˆåŠŸè·å–äº† {len(tournaments_list)} åœºé”¦æ ‡èµ›çš„æ•°æ®ï¼Œå¹¶å·²å°†å…¶å­˜å‚¨åœ¨ 'tournaments_list' å˜é‡ä¸­ã€‚")

    # --- ç¬¬äºŒæ­¥ï¼šéªŒè¯ä¸€ä¸‹åˆ—è¡¨é‡Œçš„å†…å®¹ (å¯é€‰) ---

    # æ£€æŸ¥ä¸€ä¸‹åˆ—è¡¨æ˜¯å¦ä¸ºç©º
    if tournaments_list:
        # è®©æˆ‘ä»¬æ¥çœ‹ä¸€ä¸‹åˆ—è¡¨ä¸­çš„ç¬¬ä¸€åœºé”¦æ ‡èµ›æ˜¯ä»€ä¹ˆæ ·å­çš„
        first_tournament = tournaments_list[0]
        
        print("\nè¿™æ˜¯ 'tournaments_list' ä¸­ç¬¬ä¸€åœºé”¦æ¨™è³½çš„æ•°æ®ç»“æ„ï¼š")
        print(json.dumps(first_tournament, indent=2))
        
else:
    print(f"è¯·æ±‚å¤±è´¥ï¼ŒçŠ¶æ€ç : {response.status_code}")
    print("è¯·æ£€æŸ¥ä½ çš„ç½‘ç»œè¿æ¥æˆ–APIåœ°å€æ˜¯å¦æ­£ç¡®ã€‚")
```

```python
# å‡è®¾ tournaments_list æ˜¯ä½ ä¸Šä¸€æ­¥è·å–åˆ°çš„åŒ…å«132ä¸ªé”¦æ ‡èµ›å­—å…¸çš„åˆ—è¡¨
# tournaments_list = response.json() 

# --- æ•°æ®ç­›é€‰æ­¥éª¤ ---

# 1. åˆ›å»ºä¸€ä¸ªç©ºåˆ—è¡¨ï¼Œç”¨æ¥å­˜æ”¾ç¬¦åˆæˆ‘ä»¬æ¡ä»¶çš„é”¦æ ‡èµ›ID
filtered_ids = []

print(f"å¼€å§‹ä» {len(tournaments_list)} åœºé”¦æ ‡èµ›ä¸­è¿›è¡Œç­›é€‰...")
print("ç­›é€‰æ¡ä»¶: claim_count > 3 AND claim_conflict is False\n")

# 2. éå†ä½ è·å–åˆ°çš„æ¯ä¸€åœºé”¦æ ‡èµ›æ•°æ®
for tournament in tournaments_list:
    
    # 3. æ£€æŸ¥æ¯ä¸€ä¸ªé”¦æ ‡èµ›å­—å…¸æ˜¯å¦åŒæ—¶æ»¡è¶³ä¸¤ä¸ªæ¡ä»¶
    # ä½¿ç”¨ .get() æ–¹æ³•æ¥å®‰å…¨åœ°è·å–å€¼ï¼Œå¦‚æœé”®ä¸å­˜åœ¨ï¼Œåˆ™è¿”å›ä¸€ä¸ªä¸ä¼šå¯¼è‡´é”™è¯¯çš„å€¼
    # ä¾‹å¦‚ï¼Œå¦‚æœ'claim_count'ä¸å­˜åœ¨ï¼Œæˆ‘ä»¬ç»™å®ƒä¸€ä¸ªé»˜è®¤å€¼0
    claim_count = tournament.get('claim_count', 0)
    claim_conflict = tournament.get('claim_conflict', True) # å¦‚æœ'claim_conflict'ä¸å­˜åœ¨ï¼Œé»˜è®¤å®ƒæ˜¯æœ‰é—®é¢˜çš„

    # è¿™å°±æ˜¯ç­›é€‰çš„æ ¸å¿ƒé€»è¾‘ï¼
    if claim_count > 3 and claim_conflict == False:
        # 4. å¦‚æœæ¡ä»¶æ»¡è¶³ï¼Œå°±æŠŠè¿™åœºé”¦æ ‡èµ›çš„'id'æ·»åŠ åˆ°æˆ‘ä»¬çš„åˆ—è¡¨ä¸­
        filtered_ids.append(tournament['id'])
        # æˆ‘ä»¬å¯ä»¥é¡ºä¾¿æ‰“å°å‡ºç¬¦åˆæ¡ä»¶çš„æ¯”èµ›ä¿¡æ¯ï¼Œæ–¹ä¾¿æ£€æŸ¥
        print(f"âœ… ç¬¦åˆæ¡ä»¶: ID={tournament['id']}, Title='{tournament['title']}', claim_count={claim_count}")

# 5. å¾ªç¯ç»“æŸåï¼Œæ‰“å°å‡ºæ‰€æœ‰ç¬¦åˆæ¡ä»¶çš„ID
print("\n" + "="*50)
print("ğŸ‰ ç­›é€‰å®Œæˆï¼")

if filtered_ids:
    print(f"æ€»å…±æœ‰ {len(filtered_ids)} åœºé”¦æ ‡èµ›ç¬¦åˆæ¡ä»¶ã€‚")
    print("å®ƒä»¬çš„IDåˆ†åˆ«æ˜¯:")
    print(filtered_ids)
else:
    print("æ²¡æœ‰æ‰¾åˆ°ç¬¦åˆæ¡ä»¶çš„é”¦æ ‡èµ›ã€‚")
```

```python
import time # å¯¼å…¥timeåº“ï¼Œç”¨äºæ§åˆ¶è¯·æ±‚é¢‘ç‡

# å‡è®¾ filtered_ids æ˜¯ä½ ä¸Šä¸€æ­¥ç­›é€‰åå¾—åˆ°çš„é”¦æ ‡èµ›IDåˆ—è¡¨
# ä¾‹å¦‚: filtered_ids = [5012, 5009, 4989, ...]

# 1. åˆ›å»ºä¸€ä¸ªç©ºåˆ—è¡¨ï¼Œç”¨æ¥å­˜æ”¾æˆ‘ä»¬æœ€ç»ˆæƒ³è¦çš„æ‰€æœ‰URLä¿¡æ¯
all_deck_urls = []
all_corp_deck_urls = []
all_runner_deck_urls = []

print(f"å‡†å¤‡å¼€å§‹ä» {len(filtered_ids)} åœºç­›é€‰åçš„é”¦æ ‡èµ›ä¸­è·å–ç‰Œè¡¨é“¾æ¥...\n")

# 2. éå†æ¯ä¸€ä¸ªç­›é€‰å‡ºæ¥çš„é”¦æ ‡èµ›ID
for tournament_id in filtered_ids:
    
    # æ„å»ºé’ˆå¯¹å•ä¸ªé”¦æ¨™è³½çš„API URL
    url = f"https://alwaysberunning.net/api/entries?id={tournament_id}"
    
    print(f"æ­£åœ¨è¯·æ±‚é”¦æ ‡èµ› ID: {tournament_id} ...")
    
    try:
        response = requests.get(url)
        response.raise_for_status() # æ£€æŸ¥è¯·æ±‚æ˜¯å¦æˆåŠŸ
        
        tournament_data = response.json()
        
        # 3. æ£€æŸ¥å¹¶æå– 'decklists'
        if 'decklists' in tournament_data and tournament_data['decklists']:
            
            # éå†è¿™åœºæ¯”èµ›ä¸­çš„æ¯ä¸€ä»½ç‰Œè¡¨
            for decklist in tournament_data['decklists']:
                
                # 4. æå– corp å’Œ runner çš„ç‰Œè¡¨é“¾æ¥
                corp_url = decklist.get('corp_deck_url')
                runner_url = decklist.get('runner_deck_url')
                
                # 5. å°†æå–åˆ°çš„ä¿¡æ¯ä»¥å­—å…¸çš„å½¢å¼ï¼Œå­˜å…¥æˆ‘ä»¬çš„æ€»åˆ—è¡¨ä¸­
                # è¿™æ ·åšæ›´ç»“æ„åŒ–ï¼Œæ–¹ä¾¿åç»­å¤„ç†
                if corp_url or runner_url: # åªæœ‰åœ¨è‡³å°‘æœ‰ä¸€ä¸ªURLå­˜åœ¨æ—¶æ‰æ·»åŠ 
                    all_deck_urls.append({
                        'tournament_id': tournament_id,
                        'player': decklist.get('username', 'N/A'),
                        'corp_deck_url': corp_url,
                        'runner_deck_url': runner_url
                    })
            
            print(f"  -> æˆåŠŸå¤„ç†äº† {len(tournament_data['decklists'])} ä»½ç‰Œè¡¨ã€‚")
            
        else:
            print(f"  -> è­¦å‘Š: é”¦æ ‡èµ› {tournament_id} ä¸­æ²¡æœ‰æ‰¾åˆ° 'decklists'ã€‚")

    except requests.exceptions.RequestException as e:
        print(f"  -> é”™è¯¯: è¯·æ±‚é”¦æ ‡èµ› {tournament_id} å¤±è´¥: {e}")

    # ã€é‡è¦ã€‘åšä¸€ä¸ªæœ‰ç¤¼è²Œçš„APIä½¿ç”¨è€…ï¼
    # åœ¨æ¯æ¬¡è¯·æ±‚ä¹‹é—´æš‚åœä¸€ç§’ï¼Œé¿å…å› è¯·æ±‚è¿‡äºé¢‘ç¹è€Œè¢«æœåŠ¡å™¨å±è”½ã€‚
    time.sleep(0.1)

# --- å¾ªç¯ç»“æŸåï¼ŒæŸ¥çœ‹æˆ‘ä»¬çš„æˆæœ ---
print("\n" + "="*50)
print("ğŸ‰ å…¨éƒ¨å¤„ç†å®Œæˆï¼")

if all_deck_urls:
    print(f"æ€»å…±æ”¶é›†åˆ°äº† {len(all_deck_urls)} ä»½åŒ…å«URLçš„ç‰Œè¡¨ä¿¡æ¯ã€‚")
    print("\nå‰5æ¡è®°å½•ç¤ºä¾‹ï¼š")
    # ä½¿ç”¨json.dumpså¯ä»¥æŠŠåˆ—è¡¨æ‰“å°å¾—æ›´æ•´é½
    print(json.dumps(all_deck_urls[:5], indent=2))
else:
    print("æœªèƒ½æ”¶é›†åˆ°ä»»ä½•ç‰Œè¡¨é“¾æ¥ã€‚")
```

```python
import requests
import json
import time

# å‡è®¾ filtered_ids æ˜¯ä½ ä¸Šä¸€æ­¥ç­›é€‰åå¾—åˆ°çš„é”¦æ ‡èµ›IDåˆ—è¡¨
# ä¾‹å¦‚: filtered_ids = [5051, 5020, 4989] 

# 1. åˆ›å»ºä¸¤ä¸ªç©ºåˆ—è¡¨ï¼Œåˆ†åˆ«å­˜æ”¾å…¬å¸å’Œæ½œè¢­è€…çš„URLä¿¡æ¯
all_corp_deck_urls = []
all_runner_deck_urls = []

print(f"å‡†å¤‡å¼€å§‹ä» {len(filtered_ids)} åœºç­›é€‰åçš„é”¦æ ‡èµ›ä¸­è·å–ç‰Œè¡¨é“¾æ¥...\n")

# 2. éå†æ¯ä¸€ä¸ªç­›é€‰å‡ºæ¥çš„é”¦æ ‡èµ›ID
for tournament_id in filtered_ids:
    
    # æ„å»ºé’ˆå¯¹å•ä¸ªé”¦æ ‡èµ›çš„API URLå’Œå‚æ•°
    params = {'id': tournament_id}
    url = "https://alwaysberunning.net/api/entries"
    
    print(f"æ­£åœ¨è¯·æ±‚é”¦æ ‡èµ› ID: {tournament_id} çš„å‚èµ›åˆ—è¡¨...")
    
    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        
        # 3. response.json() è¿”å›çš„æ˜¯ä¸€ä¸ªå‚èµ›è€…åˆ—è¡¨ (list of entries)
        # æˆ‘ä»¬ç›´æ¥æŠŠå®ƒå‘½åä¸º entries_list
        entries_list = response.json()
        
        # 4. æ£€æŸ¥è¿™ä¸ªå‚èµ›è€…åˆ—è¡¨æ˜¯å¦ä¸ºç©º
        if entries_list:
            print(f"  -> æˆåŠŸè·å–äº† {len(entries_list)} æ¡å‚èµ›è®°å½•ã€‚")
            
            # 5. éå†è¿™åœºæ¯”èµ›ä¸­çš„ã€æ¯ä¸€æ¡å‚èµ›è®°å½•ã€‘(æ¯ä¸€ä¸ªå‚èµ›è€…å­—å…¸)
            for entry in entries_list:
                
                # 6. ä»æ¯ä¸ªå‚èµ›è€…(entry)å­—å…¸ä¸­æå–URL
                corp_url = entry.get('corp_deck_url')
                runner_url = entry.get('runner_deck_url')
                player_name = entry.get('user_name', 'N/A')
                
                # 7. å¦‚æœå…¬å¸URLå­˜åœ¨ï¼Œå°±æŠŠå®ƒå­˜åˆ°å…¬å¸åˆ—è¡¨ä¸­
                if corp_url:
                    all_corp_deck_urls.append({
                        'tournament_id': tournament_id,
                        'player': player_name,
                        'corp_deck_url': corp_url
                    })
                
                # 8. å¦‚æœæ½œè¢­è€…URLå­˜åœ¨ï¼Œå°±æŠŠå®ƒå­˜åˆ°æ½œè¢­è€…åˆ—è¡¨ä¸­
                if runner_url:
                    all_runner_deck_urls.append({
                        'tournament_id': tournament_id,
                        'player': player_name,
                        'runner_deck_url': runner_url
                    })
        else:
            print(f"  -> é”¦æ ‡èµ› {tournament_id} æ²¡æœ‰è¿”å›å‚èµ›è®°å½•ã€‚")

    except requests.exceptions.RequestException as e:
        print(f"  -> é”™è¯¯: è¯·æ±‚é”¦æ ‡èµ› {tournament_id} å¤±è´¥: {e}")

    # ç¤¼è²Œåœ°æš‚åœä¸€ç§’
    time.sleep(1)

# --- å¾ªç¯ç»“æŸåï¼ŒæŸ¥çœ‹æˆ‘ä»¬çš„æˆæœ ---
print("\n" + "="*50)
print("ğŸ‰ å…¨éƒ¨å¤„ç†å®Œæˆï¼")

# å…¬å¸URLè¾“å‡º
if all_corp_deck_urls:
    print(f"\næ€»å…±æ”¶é›†åˆ°äº† {len(all_corp_deck_urls)} ä»½å…¬å¸ç‰Œè¡¨é“¾æ¥ã€‚")
    print("å‰5æ¡è®°å½•ç¤ºä¾‹ï¼š")
    print(json.dumps(all_corp_deck_urls[:5], indent=2))
else:
    print("\næœªèƒ½æ”¶é›†åˆ°ä»»ä½•å…¬å¸ç‰Œè¡¨é“¾æ¥ã€‚")

# æ½œè¢­è€…URLè¾“å‡º
if all_runner_deck_urls:
    print(f"\næ€»å…±æ”¶é›†åˆ°äº† {len(all_runner_deck_urls)} ä»½æ½œè¢­è€…ç‰Œè¡¨é“¾æ¥ã€‚")
    print("å‰5æ¡è®°å½•ç¤ºä¾‹ï¼š")
    print(json.dumps(all_runner_deck_urls[:5], indent=2))
else:
    print("\næœªèƒ½æ”¶é›†åˆ°ä»»ä½•æ½œè¢­è€…ç‰Œè¡¨é“¾æ¥ã€‚")
    
```

```python
# --- ç¬¬ä¸€æ­¥ï¼šå®šä¹‰ç›®æ ‡ ---

# æˆ‘ä»¬æƒ³è¦è·å–çš„å¡ç»„ID
decklist_id = 91005

# ä½¿ç”¨ f-string æ„å»ºå®Œæ•´çš„API URL
api_url = f"https://netrunnerdb.com/api/2.0/public/decklist/{decklist_id}"

print(f"å‡†å¤‡è¯·æ±‚NetrunnerDB APIï¼Œç›®æ ‡åœ°å€ï¼š\n{api_url}\n")


# --- ç¬¬äºŒæ­¥ï¼šå‘é€è¯·æ±‚å¹¶å¤„ç†å“åº” ---

try:
    response = requests.get(api_url)
    response.raise_for_status() # æ£€æŸ¥æ˜¯å¦æœ‰HTTPé”™è¯¯ (å¦‚ 404 Not Found)

    print("âœ… è¯·æ±‚æˆåŠŸï¼æœåŠ¡å™¨å·²è¿”å›æ•°æ®ã€‚\n")

    # å°†è¿”å›çš„JSONæ•°æ®è§£ææˆPythonå­—å…¸
    decklist_data = response.json()

    # --- ç¬¬ä¸‰æ­¥ï¼šæ‰“å°å¹¶åˆ†æè¿”å›çš„æ•°æ®ç»“æ„ ---

    print("="*50)
    print("ä»¥ä¸‹æ˜¯APIè¿”å›çš„å®Œæ•´æ•°æ®ç»“æ„ï¼š")
    # ä½¿ç”¨ json.dumps å¹¶è®¾ç½® indent=2 å¯ä»¥è®©JSONæ•°æ®æ ¼å¼åŒ–è¾“å‡ºï¼Œéå¸¸æ˜“è¯»
    print(json.dumps(decklist_data, indent=2))
    print("="*50)

    # --- ç¬¬å››æ­¥ï¼šæå–æ ¸å¿ƒä¿¡æ¯ ---

    # æ ¹æ®APIæ–‡æ¡£å’Œä¸Šé¢çš„è¾“å‡ºï¼Œæˆ‘ä»¬å‘ç°æ•°æ®åŒ…è£¹åœ¨ "data" é”®çš„åˆ—è¡¨é‡Œ
    if decklist_data.get('success') and decklist_data.get('data'):
        # è·å–å¡ç»„ä¿¡æ¯å­—å…¸ï¼Œå®ƒåœ¨dataåˆ—è¡¨çš„ç¬¬ä¸€ä¸ªä½ç½®
        deck_info = decklist_data['data'][0]
        
        # æˆ‘ä»¬æœ€å…³å¿ƒçš„å°±æ˜¯ 'cards' è¿™ä¸ªé”®ï¼Œå®ƒåŒ…å«äº†æ‰€æœ‰çš„å¡ç‰Œ
        cards = deck_info.get('cards')
        deck_name = deck_info.get('name')

        print(f"\næˆåŠŸè§£æå¡ç»„ '{deck_name}'ï¼")
        
        if cards:
            print("\nå¡ç»„ä¸­çš„'cards'éƒ¨åˆ†åŒ…å«ä»¥ä¸‹å†…å®¹ï¼ˆé”®ä¸ºå¡ç‰ŒIDï¼Œå€¼ä¸ºæ•°é‡ï¼‰:")
            # 'cards' æœ¬èº«æ˜¯ä¸€ä¸ªå­—å…¸ï¼Œæˆ‘ä»¬æ‰“å°å‡ºæ¥çœ‹çœ‹
            print(cards)
            
            # è®¡ç®—å¡ç»„æ€»å¼ æ•°
            total_cards = sum(cards.values())
            print(f"\nå¡ç»„æ€»å¼ æ•°: {total_cards}")
            
        else:
            print("æœªèƒ½åœ¨æ­¤å¡ç»„ä¿¡æ¯ä¸­æ‰¾åˆ°'cards'é”®ã€‚")
            
    else:
        print("APIè¿”å›çš„æ•°æ®æ ¼å¼ä¸ç¬¦åˆé¢„æœŸï¼Œæˆ–è€…è¯·æ±‚æœªæˆåŠŸã€‚")

except requests.exceptions.RequestException as e:
    print(f"âŒ è¯·æ±‚å¤±è´¥: {e}")
```

```python
import pandas as pd # å¯¼å…¥pandasåº“ï¼Œé€šå¸¸ç®€å†™ä¸ºpd

# å‡è®¾ all_corp_deck_urls å’Œ all_runner_deck_urls æ˜¯ä½ å·²ç»æˆåŠŸè·å–çš„åˆ—è¡¨
# all_corp_deck_urls = [{'tournament_id': 5051, 'player': 'rubenpieters', 'corp_deck_url': 'https://netrunnerdb.com/en/decklist/91006/...'}, ...]

# --- ç¬¬ä¸€éƒ¨åˆ†ï¼šå¤„ç†å…¬å¸æ–¹å¡ç»„ ---

# 1. åˆ›å»ºä¸€ä¸ªæœ€ç»ˆçš„ã€å¤§çš„åˆ—è¡¨ï¼Œç”¨æ¥å­˜æ”¾æ¯ä¸€æ¡å¡ç‰Œè®°å½•
corp_card_data = []

print(f"å‡†å¤‡å¤„ç† {len(all_corp_deck_urls)} ä¸ªå…¬å¸æ–¹å¡ç»„é“¾æ¥...\n")

# 2. éå†æˆ‘ä»¬ä¹‹å‰æ”¶é›†åˆ°çš„æ¯ä¸€ä¸ªåŒ…å«URLçš„å­—å…¸
for deck_info in all_corp_deck_urls:
    
    url = deck_info['corp_deck_url']
    
    # å¦‚æœURLä¸ºç©ºï¼Œåˆ™è·³è¿‡æ­¤æ¬¡å¾ªç¯
    if not url:
        continue

    try:
        # 3. ä»URLä¸­è§£æå‡º decklist_id (è¿™æ˜¯å…³é”®ä¸€æ­¥ï¼)
        # ä¾‹å¦‚: 'https://netrunnerdb.com/en/decklist/91006/...' -> '91006'
        decklist_id = url.split('/')[5]
        
        # 4. ä½¿ç”¨IDè°ƒç”¨NetrunnerDB API
        api_url = f"https://netrunnerdb.com/api/2.0/public/decklist/{decklist_id}"
        print(f"æ­£åœ¨è¯·æ±‚ Decklist ID: {decklist_id} ...")
        
        response = requests.get(api_url)
        response.raise_for_status()
        decklist_data = response.json()
        
        # 5. æå– 'cards' å­—å…¸
        if decklist_data.get('success') and decklist_data.get('data'):
            cards_dict = decklist_data['data'][0].get('cards')
            
            if cards_dict:
                # 6. ã€æ ¸å¿ƒã€‘éå† 'cards' å­—å…¸ï¼Œå°†æ¯å¼ å¡ä½œä¸ºä¸€æ¡æ–°è®°å½•æ·»åŠ 
                for card_id, quantity in cards_dict.items():
                    corp_card_data.append({
                        'tournament_id': deck_info['tournament_id'],
                        'player': deck_info['player'],
                        'decklist_id': decklist_id,
                        'card_id': card_id,
                        'quantity': quantity
                    })
        
        # ç¤¼è²Œåœ°æš‚åœ
        time.sleep(1)

    except Exception as e:
        print(f"  -> å¤„ç†URL {url} æ—¶å‘ç”Ÿé”™è¯¯: {e}")
        continue # å³ä½¿å•ä¸ªURLå‡ºé”™ï¼Œä¹Ÿç»§ç»­å¤„ç†ä¸‹ä¸€ä¸ª

# --- ç¬¬äºŒéƒ¨åˆ†ï¼šå°†æ”¶é›†åˆ°çš„æ•°æ®è½¬æ¢æˆè¡¨æ ¼ (Pandas DataFrame) ---
print("\n" + "="*50)
print("ğŸ‰ å…¬å¸æ–¹æ•°æ®å¤„ç†å®Œæˆï¼æ­£åœ¨åˆ›å»ºè¡¨æ ¼...")

if corp_card_data:
    # è¿™å°±æ˜¯â€œæ‹¼æˆè¡¨æ ¼â€çš„é­”æ³•ï¼
    corp_df = pd.DataFrame(corp_card_data)

    print("\nå…¬å¸æ–¹å¡ç‰Œæ•°æ®è¡¨æ ¼åˆ›å»ºæˆåŠŸï¼")
    print("è¡¨æ ¼ä¿¡æ¯æ¦‚è§ˆï¼š")
    corp_df.info() # æ˜¾ç¤ºè¡¨æ ¼çš„ç»“æ„ä¿¡æ¯

    print("\nè¡¨æ ¼å‰5è¡Œé¢„è§ˆï¼š")
    print(corp_df.head()) # æ˜¾ç¤ºè¡¨æ ¼çš„å‰5è¡Œ
else:
    print("æœªèƒ½æ”¶é›†åˆ°ä»»ä½•å…¬å¸æ–¹å¡ç‰Œæ•°æ®ã€‚")
```
